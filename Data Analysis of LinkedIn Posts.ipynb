# Analyzing the words in the posts

This analysis is to see what words I use frequently in my posts and what words are mostly used in posts with the most engagement.

### Analyzing the words in the top 10 posts

```python
dfToList = df['postContent'].tolist()
```

```python
df_top_likes = df.sort_values(by=['likeCount'], ascending=False).head(10)
```

```python
top_likes = df_top_likes['postContent'].tolist()
```

```python
t = []

for i in top_likes:
    if (type(i)) == float:
        i = str(i)
    t.append((i).split())
```

```python
d = []
for i in t:
    for b in i:
        d.append(b)
```

```python
# I could write a counter function but I prefer to use the collection module
from collections import Counter

word_count = Counter(d)

print(word_count.most_common())
```
